```python
import pandas as pd
import numpy as np
```

```python
# append = 행 합침
# merge = 열 합침

df1 = pd.DataFrame({'고객번호 ' : [1001, 1002, 1003, 1004, 1005, 1006, 1007], 
                '이름' : ['둘리', '도우너', '또치', '고길동', '희동', '마이콜', '영희']
                 })
df1

# 피벗 테이블

```

```python
df2 = pd.DataFrame({'고객번호 ' : [1001, 1001, 1005, 1006, 1008, 1001], 
                '금액' : [10000, 2000, 15000, 4000, 5000, 30000]
                 })
df2
```

```python
# 컬럼으로 합치는 방법
# merge
pd.merge(df1, df2) # 고객 번호가 존재하는 값을 가졌을 경우에만
```

```python
#merge는 합치는 컬럼에 영향을 미치지않는다
df1 
```

```python
# 그래서 값을 다른 값에 저장해서 사용하면 된다
#how = 'inner'이 기본값 / outer를 사용하면 적용되지않은 값도 표현
df3 = pd.merge(df1, df2, how='outer') 
df3
```

```python
# left = df1 기준 존재하는 값 모두 표기
pd.merge(df1, df2, how='left')
```

```python
# right = df2 기준 존재하는 값 모두 표기
pd.merge(df1, df2, how='right')
```

```python
# NaN 처리 
df4 = pd.merge(df1, df2, how='outer')
df4
#df4의 null정보 확인 1
df4.info()
```

10개의 인덱스중
이름은 null이 1개
금액은 null이 4개

```python
# True값으로 null값 확인 2 
df4.isnull()
```

```python
# fillna = null값을 0으로 바꿈
df5 = df4.fillna(0)
df5
```

```python
# df4에서 금액만 0으로 바꾸어라
df5 = df4['금액'].fillna(0)
df5
```

```python
# df4에서 이름만 0으로 바꾸어라
df6 = df4['이름'].fillna('--')
df5
```

```python
# dropna 는 null값을 제거하고 표기
df4.dropna()
```

```python
# 특정 column을 골라서 null값 제거 가능
df4['금액'].dropna()
```

```python
# dropna 는 기본적으로 행을 기준으로 삭제하는것
# axis =1 을 하면 열 삭제
df4.dropna(axis=1)
```

```python
# 널값 확인 방법 3
df4.isnull().sum()
```
